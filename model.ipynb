```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LinearRegression

# Sample labeled data
data = pd.DataFrame({
    'text': [
        "I love this movie! It's fantastic!",
        "This product is terrible. I'm very disappointed.",
        "The weather today is okay, nothing special.",
        "I feel so happy when I'm with my friends.",
        "I'm feeling sad because of recent events."
    ],
    'intensity': [0.9, 0.1, 0.5, 0.8, 0.2]  # Labeled intensity scores (ranging from 0 to 1)
})

# Split the data into input features (text) and target labels (intensity scores)
X = data['text']
y = data['intensity']

# Vectorize the text data using CountVectorizer
vectorizer = CountVectorizer()
X_vectorized = vectorizer.fit_transform(X)

# Train a Linear Regression model
regression_model = LinearRegression()
regression_model.fit(X_vectorized, y)

# New text data for prediction
new_text = ["I'm feeling ecstatic today!", "I'm feeling a bit down."]

# Vectorize the new text data
new_text_vectorized = vectorizer.transform(new_text)

# Predict intensity scores for the new text data
predicted_intensity_scores = regression_model.predict(new_text_vectorized)

print("Predicted Intensity Scores:", predicted_intensity_scores)

```

```python
import pandas as pd
from textblob import TextBlob
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical

# Load the dataset
data = pd.read_csv('E:\\Emolnt\\trainingtrans.csv', encoding='latin-1')

# Purely statistical model using TextBlob
def textblob_emotion_intensity(text):
    analysis = TextBlob(text)
    polarity = analysis.sentiment.polarity
    subjectivity = analysis.sentiment.subjectivity
    return polarity, subjectivity

data['polarity'], data['subjectivity'] = zip(*data['text'].map(textblob_emotion_intensity))
print(data.head())

# Deep learning model
# Prepare the text data
tokenizer = Tokenizer(num_words=5000, split=' ')
tokenizer.fit_on_texts(data['text'])

X = tokenizer.texts_to_sequences(data['text'])
X = pad_sequences(X)

# Prepare the labels
y = pd.get_dummies(data['polarity']).values

# Define the model architecture
model = Sequential()
model.add(Dense(512, activation='relu', input_shape=(X.shape[1],)))
model.add(Dropout(0.2))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(len(y[0]), activation='softmax'))

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer=Adam(learning_rate=0.001), metrics=['accuracy'])

# Train the model
model.fit(X, y, batch_size=32, epochs=10, validation_split=0.2)

# Save the model (optional)
model.save('emotion_intensity_model.h5')
```

```python
def polarity_to_intensity(polarity):
    """Map polarity scores to intensity scores."""
    if polarity < 0:
        # Negative polarity corresponds to negative emotions
        intensity = -polarity
    else:
        # Positive polarity corresponds to positive emotions
        intensity = polarity

    # Scale the intensity to a range of 0 to 1
    intensity = intensity / 1.0

    return intensity

# Assume `data` is your DataFrame and `polarity` is the column with polarity scores
data['intensity'] = data['polarity'].apply(polarity_to_intensity)
```

```python
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Sample data for sentiment analysis
data = ["I love this product", "This is the worst thing I have ever bought"]

# Initialize TextBlob and VADER models
textblob_model = TextBlob("")
vader_model = SentimentIntensityAnalyzer()

# Function to perform sentiment analysis using TextBlob
def analyze_textblob(text):
    analysis = TextBlob(text)
    polarity = analysis.sentiment.polarity
    subjectivity = analysis.sentiment.subjectivity
    return polarity, subjectivity

# Function to perform sentiment analysis using VADER
def analyze_vader(text):
    analysis = vader_model.polarity_scores(text)
    polarity = analysis['compound']
    subjectivity = analysis['compound'] # VADER doesn't have a subjectivity score, so we'll use polarity as an approximation
    return polarity, subjectivity

# Perform sentiment analysis on sample data using both models
textblob_polarities = [analyze_textblob(text)[0] for text in data]
vader_polarities = [analyze_vader(text)[0] for text in data]

# Print results
print("TextBlob Polarities:", textblob_polarities)
print("VADER Polarities:", vader_polarities)

# Calculate accuracy of each model
textblob_correct = sum(1 for text, truth in zip(textblob_polarities, vader_polarities) if text > 0 == truth)
vader_correct = sum(1 for text, truth in zip(textblob_polarities, vader_polarities) if text == truth)

textblob_accuracy = textblob_correct / len(data)
vader_accuracy = vader_correct / len(data)

print("TextBlob Accuracy:", textblob_accuracy)
print("VADER Accuracy:", vader_accuracy)
```

```python
import pandas as pd
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from sklearn.metrics import f1_score

# Load data from CSV file
data = pd.read_csv("E:\\Emolnt\\trainingtrans.csv")

# Initialize TextBlob and VADER models
textblob_model = TextBlob("")
vader_model = SentimentIntensityAnalyzer()

# Function to perform sentiment analysis using TextBlob
def analyze_textblob(text):
    analysis = TextBlob(text)
    polarity = analysis.sentiment.polarity
    subjectivity = analysis.sentiment.subjectivity
    return polarity, subjectivity

# Function to perform sentiment analysis using VADER
def analyze_vader(text):
    analysis = vader_model.polarity_scores(text)
    polarity = analysis['compound']
    subjectivity = analysis['compound'] # VADER doesn't have a subjectivity score, so we'll use polarity as an approximation
    return polarity, subjectivity

# Perform sentiment analysis on the data using both models
textblob_polarities = [analyze_textblob(text)[0] for text in data['text']]
vader_polarities = [analyze_vader(text)[0] for text in data['text']]

# Define the ground truth labels
ground_truth = [1 if label == "positive" else -1 for label in data['text']]

# Calculate F1 score for TextBlob
textblob_f1 = f1_score(ground_truth, [1 if polarity > 0 else -1 for polarity in textblob_polarities])

# Calculate F1 score for VADER
vader_f1 = f1_score(ground_truth, [1 if polarity > 0 else -1 for polarity in vader_polarities])

# Print the F1 scores for both models
print("TextBlob F1 Score:", textblob_f1)
print("VADER F1 Score:", vader_f1)

# Determine which model has a higher F1 score
if textblob_f1 > vader_f1:
    print("TextBlob has a higher F1 score.")
elif vader_f1 > textblob_f1:
    print("VADER has a higher F1 score.")
else:
    print("Both models have the same F1 score.")
```
